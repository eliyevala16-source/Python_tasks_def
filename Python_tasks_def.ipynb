```
def sum_tek_or_cut():
  sum=0
  while True:
    eded=int(input('eded daxil edin: '))
    if eded == -1:
      break
    if eded > 0:
      sum += eded
  if sum % 2 == 0:
    print('ededlerin cemi cutdur')
  else:
    print('ededlerin cemi tekdir')
sum_tek_or_cut()
```

```
def quotient_tek_or_cut():
  a = int(input('eded daxil edin: '))
  b = int(input('eded daxil edin: '))
  if b == 0:
    print('b 0 ola bilmez')
  else:
    quotient = a // b
    if quotient % 2 == 0:
      print('quotient cutdur')
    else:
      print('False')
quotient_tek_or_cut()
```

```
def power():
  n = int(input('eded n daxil edin: '))
  k = int(input('eded k daxil edin: '))
  if k ** k == n:
    print('True')
  else:
    print('False')
power()
```

```
def pronic_or_heteromecic():
   p = int(input('eded daxil edin: '))
   n = 0
   while n * (n + 1)  <= p:
    if n * (n + 1) == p:
      print('pronic')
      return
    n += 1
   print('heteromecic')
pronic_or_heteromecic()
```

```
def reqem_sayi():
  eded = int(input('eded daxil edin: '))
  eded = abs(eded)
  say = 0
  if eded == 0:
    print('reqemlerinin sayi: 1')
    return
  while  eded > 0:
    say += 1
    eded //= 10
  print(f'reqemlerinin sayi: {say}')
reqem_sayi()

```

```
def disarium_or_not():
    N = int(input('eded daxil edin: '))
    index = 0
    temp = N
    while temp > 0:
        temp //= 10
        index += 1
    N2 = N
    sum = 0
    while N2 > 0:
        eded = N2 % 10
        sum += eded ** index
        index -= 1
        N2 //= 10

    if N == sum:
        print('disarium')
    else:
        print('not disarium')

disarium_or_not()



```

```
def curzon_or_not():
  n = int(input('eded daxil edin: '))
  if (2 ** n + 1) % (2 * n + 1) == 0:
    print('curzon')
  else:
    print('not curzon')
curzon_or_not()
```

```
def kempner():
  n = int(input('eded daxil edin: '))
  i = 1
  fac = 1
  while True:
    fac *= i
    if fac % n == 0:
      print(f'{n} -> {i}!')
      break
    i += 1
kempner()
```

```
def moran():
  n = int(input('eded daxil edin:'))
  temp = n
  sum = 0
  while temp > 0:
    sum += temp % 10
    temp //= 10
  if n % sum != 0:
    print('not moran')
    return
  quotient = n // sum
  if quotient < 2:
    print('not moran')
    return
  i = 2
  while i * i <= quotient:
    if quotient % i == 0:
      print('not moran')
      return
    i += 1
  print('moran')
moran()


```

```
def hiper_sade():
    n = int(input("eded daxil edin: "))
    eded = n
    while eded > 0:
        if eded < 2:
            print("Hiper-sade deyil")
            return
        for i in range(2, int(eded ** 0.5) + 1):
            if eded % i == 0:
                print("Hiper-sade deyil")
                return
        eded //= 10
    print("Hiper-sadedir")
hiper_sade()
```
